```python
# Import libraries
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv('../../data/preprocessed/hba1c_classification/us_hba1c_range_dataset.csv')
# df = pd.read_csv('../../data/preprocessed/hba1c_classification/uk_hba1c_range_dataset.csv')
```

```python
# Data inspection

print("First ten rows of the dataset:")
display(df.head(10))

print("\nDataset dimensions:")
print(f"Number of rows: {df.shape[0]}, Number of columns: {df.shape[1]}")

print("\nDataset information:")
df.info()

print("\nStatistical summary:")
display(df.describe())
```

```python
# Check for unique elements

for column in df.columns:
    print(f"Unique values in '{column}':")
    print(df[column].unique())
    print("\n")
```

```python
# Check for value count

for column in df.columns:
    print(f"Unique values in '{column}':")
    print(df[column].value_counts())
    print("\n")
```

```python
# One hot encoding for categorical features and target
df = pd.get_dummies(df, columns = ['gender', 'smoking_history', 'diabetes_status'])
```

```python
# drop origianl diabetes column
df = df.drop(columns=['diabetes'])
```

```python
# Data re-inspection

print("First ten rows of the dataset:")
display(df.head(10))

print("\nDataset dimensions:")
print(f"Number of rows: {df.shape[0]}, Number of columns: {df.shape[1]}")

print("\nDataset information:")
df.info()

print("\nStatistical summary:")
display(df.describe())
```

```python
# Check for unique elements

for column in df.columns:
    print(f"Unique values in '{column}':")
    print(df[column].unique())
    print("\n")
```

```python
# Check for value count

for column in df.columns:
    print(f"Unique values in '{column}':")
    print(df[column].value_counts())
    print("\n")
```

```python
# Save encoded dataset

df.to_csv('../../data/encoded/one_hot_target/encoded_us_hba1c_range_dataset_one_hot_target.csv', index=False)
# df.to_csv('../../data/encoded/one_hot_target/encoded_uk_hba1c_range_dataset_one_hot_target.csv', index=False)
```
